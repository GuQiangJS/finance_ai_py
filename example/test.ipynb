```python
close=pd.to_numeric(df['close'])
```

```python
import matplotlib.ticker as ticker
```

```python
ax=sns.lineplot(data=close)
ax.xaxis.set_major_locator(ticker.AutoLocator())
```

```python
import talib
```

```python
sma=talib.SMA(close)
```

```python
sma=pd.DataFrame({'原始值':close,'移动平均线':sma})
```

```python
ax=sns.lineplot(data=sma)
ax.xaxis.set_major_locator(ticker.AutoLocator())
```

```python
from talib import MA_Type
upper, middle, lower = talib.BBANDS(close, matype=MA_Type.T3)
bbands=sma=pd.DataFrame({'原始值':close,'均值':middle,'上线':upper,'下线':lower})
```

```python
plt.figure(figsize=(20,10))
ax=sns.lineplot(data=bbands[:100])
ax.xaxis.set_major_locator(ticker.AutoLocator())
```
