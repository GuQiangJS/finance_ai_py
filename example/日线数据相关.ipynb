```python
columns=['open','close','high','low','volume']
```

```python
df_all=df_tushare[columns].join(df_qfq[columns],rsuffix='_qfq').join(df_hfq[columns],rsuffix='_hfq')
df_all.head()
```

```python
#计算日收益
df_all['daily_return']=df_all['close']/df_all['close'].shift(1)-1
df_all['daily_return_qfq']=df_all['close_qfq']/df_all['close_qfq'].shift(1)-1
df_all['daily_return_hfq']=df_all['close_hfq']/df_all['close_hfq'].shift(1)-1
```

```python
#无论如何复权，日收益是一样的
df_all.head()
```

标准化

optInNbDevUp和optInNbDevDn是计算上下波段的移动平均线上下的标准偏差数。

optInMAType是要使用的移动平均线类型，
即SMA  - 简单移动平均线
EMA  - 指数平均
WMA  - 加权移动平均线
DEMA  - 双指数移动平均线
TEMA  - 三倍指数移动平均线
TRIMA  - 三角移动平均线
KAMA  - 考夫曼自适应移动平均线
MAMA  -  MESA自适应移动平均线
T3  - 三次指数移动平均线，例如。

upper，middle，lower = talib.BBANDS（close，matype = MA_Type.T3）P。

```python
def bbands_plot(series,t=20,matype=MA_Type.WMA):
    """绘制布林带
    默认为20日均线，加权计算
    """
    bbands_up,bbands_mid,bbands_low=QA.QAIndicator.talib_series.BBANDS(series,t,matype=matype)
    bbands_mid.plot()
    bbands_up.plot()
    bbands_low.plot()
```

```python
bbands_plot(df_tushare['close'])
```
