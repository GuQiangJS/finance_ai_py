```python
import warnings
warnings.filterwarnings('ignore')
import QUANTAXIS as QA
import talib
import jupyter_helper
import numpy as np
from pyecharts import Line

import os
os.path.abspath(os.path.dirname(os.getcwd()))
import sys
sys.path.append(os.path.abspath(os.path.dirname(os.getcwd())))

from finance_ai_py import quantaxis_ext
from talib import MA_Type
```

```python
s,e=jupyter_helper.get_start_end_date()
benchmark_code=jupyter_helper.get_benchmark_code()
stock_code='600436'
timeperiod=20
nbdevup=2
nbdevdn=2
INIT_CASH=10000
UP=70
LOW=30
MATYPE=MA_Type.EMA
```

```python
stock_data=QA.QA_fetch_stock_day_adv(stock_code,s,e)
```

```python
from talib import MA_Type
df=stock_data.data
df['qfq_close']=stock_data.to_qfq().data['close']
df=df.reset_index().set_index('date')
df['DAILY_RETURN']=np.round(jupyter_helper.calc_daily_return(df['close']),4)
df['MA_UPPER'],df['MA_MIDDLE'],df['MA_LOWER']=talib.BBANDS(np.array(df['qfq_close']), timeperiod=timeperiod, nbdevup=nbdevup, nbdevdn=nbdevdn,matype=MATYPE)
df['RSI']=talib.RSI(np.array(df['qfq_close']))
df=df.dropna()
```

```python
df
```

## 绘制日线（前复权）

```python
line = Line('')
markpoint_b=[]
markpoint_s=[]
for x in df[(df['qfq_close']<df['MA_LOWER']) & (df['RSI']<LOW)].index:
    markpoint_b.append({"coord":[x.date(),np.round(df.loc[x]['qfq_close'],2)],'name':'buy-{}'.format(x.date())})
for x in df[(df['qfq_close']>df['MA_UPPER']) & (df['RSI']>UP)].index:
    markpoint_s.append({"coord":[x.date(),np.round(df.loc[x]['qfq_close'],2)],'name':'sale-{}'.format(x.date())})
# 通过使用同一日线数据但是根据买入/卖出分别绘制的方式，实现不同颜色标记买入和卖出点
line.add('qfq_close_b', df.index.date, np.round(df['qfq_close'], 2),datazoom_extra_type='both',
                 is_datazoom_extra_show=True,
                 datazoom_extra_orient='horizontal',
        mark_point=markpoint_b)
line.add('qfq_close_s', df.index.date, np.round(df['qfq_close'], 2),datazoom_extra_type='both',
                 is_datazoom_extra_show=True,
                 datazoom_extra_orient='horizontal',
        mark_point=markpoint_s)
line.add('MA_LOWER', df.index.date, np.round(df['MA_LOWER'], 2),datazoom_extra_type='both',
                 is_datazoom_extra_show=True,
                 datazoom_extra_orient='horizontal')
line.add('MA_UPPER', df.index.date, np.round(df['MA_UPPER'], 2),datazoom_extra_type='both',
                 is_datazoom_extra_show=True,
                 datazoom_extra_orient='horizontal')
line.add('MA_MIDDLE', df.index.date, np.round(df['MA_MIDDLE'], 2),datazoom_extra_type='both',
                 is_datazoom_extra_show=True,
                 datazoom_extra_orient='horizontal')
line
```

## 模拟下单

```python
stock_account=quantaxis_ext.QA_Account_Ext(init_cash=INIT_CASH)
amount=100
Broker = QA.QA_BacktestBroker()

for index in df.index:
    if df.loc[index]['qfq_close']<df.loc[index]['MA_LOWER'] and df.loc[index]['RSI']<LOW:
        order = stock_account.send_order(
                    code=stock_code,
                    time=index.date(),
                    amount=amount,
                    towards=QA.ORDER_DIRECTION.BUY,
                    price=0,
                    order_model=QA.ORDER_MODEL.CLOSE,
                    amount_model=QA.AMOUNT_MODEL.BY_AMOUNT  # 按数量下单
                    # 在按金额下单的时候,应给予 money参数
                    # 在按数量下单的时候,应给予 amount参数
                )
                # print(item.to_json()[0])
        Broker.receive_order(QA.QA_Event(order=order))

        trade_mes = Broker.query_orders(stock_account.account_cookie, 'filled')
        res = trade_mes.loc[order.account_cookie, order.realorder_id]
        order.trade(res.trade_id, res.trade_price, res.trade_amount,res.trade_time) 
    elif df.loc[index]['qfq_close']>df.loc[index]['MA_UPPER'] and df.loc[index]['RSI']>UP:
#         if stock_account.sell_available.get(stock_code, 0) >= amount:#当前有可卖的数量时
        if stock_account.sell_available.get(stock_code, 0) >= amount and df.loc[index]['qfq_close']>stock_account.hold_price_fifo().at[stock_code][0]:
            order = stock_account.send_order(
                        code=stock_code,
                        time=index.date(),
                        amount=amount,
#                         amount=stock_account.sell_available.get(stock_code, 0),
                        towards=QA.ORDER_DIRECTION.SELL,
                        price=0,
                        order_model=QA.ORDER_MODEL.CLOSE,
                        amount_model=QA.AMOUNT_MODEL.BY_AMOUNT  # 按数量下单
                        # 在按金额下单的时候,应给予 money参数
                        # 在按数量下单的时候,应给予 amount参数
                    )
                    # print(item.to_json()[0])
            Broker.receive_order(QA.QA_Event(order=order))

            trade_mes = Broker.query_orders(stock_account.account_cookie, 'filled')
            res = trade_mes.loc[order.account_cookie, order.realorder_id]
            order.trade(res.trade_id, res.trade_price, res.trade_amount,res.trade_time)
    stock_account.settle()
```
